- IF -ELIF -ELSE

- I'm telling to the computer ""Hey if this case happens, perform some action ,else if another case happens , perform some other action  , else if None of the above cases happend, perform this action"

if case1 :

  perform action 1

elif case2:

   perform action 2

else :

  perform action 3 

```python
# single condition

a= 2

if a<10:                # it is true
    print("yes")
```

```python
# single condition

a= 2

if a>10:                # if  it is false , we don't get any output
    print("yes")
```
