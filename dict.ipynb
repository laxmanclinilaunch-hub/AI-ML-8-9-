- keys()

- it returns the list of keys in the dict object



```python
marvel_dict.keys()
```

```python
list(marvel_dict.keys())
```

- values()

- - it returns the list of values  in the dict object

```python
marvel_dict.values()
```

```python
list(marvel_dict.values())
```

```python
tuple(marvel_dict.values())
```

- items()

```python
marvel_dict.items()
```

```python
list(marvel_dict.items())
```
