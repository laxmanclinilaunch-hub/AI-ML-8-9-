# - How to create empty str, empty list , empty tuples, empty sets, empty dict ?

```python
p= " "


print (f" this is my empty str : {p}")


q= [ ]

print(f"this is my emptylist :{q}")


r= ( )

print(f"this is my empty tuples :{r}")



s= { }


print (f" this is my empty dict : {s}")


z= set()

print (f" this is my empty set : {z}")
```
