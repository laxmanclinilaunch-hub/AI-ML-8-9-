- Zip

```python
city_list=["Hyderabad","mumbai","kochi","chennai","thane"]


river_list=["musi","mithi","kavari","adayar","upan"]
```

```python
zip(city_list,river_list)    # saved only in memory
```

```python
list(zip(city_list,river_list) )
```

```python
tuple(zip(city_list,river_list) )
```
