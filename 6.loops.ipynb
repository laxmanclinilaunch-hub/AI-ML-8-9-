- two types

         - it helps to excute the block of code repetedly until completion of iterable objects

    - while/ indefinte loop:
        - it helps to excute the block of code repetedly until  the has True 

```python
a ="we are  working healthcareprofessionals"


for i in a:
    print(i)
```

```python
b=[1,2,5,90,'divya','arunachalam',7+8j,True,False]

for i in b:
    print(i)
```

- Break & continue & Pass

# Break
 -  when specific condition met, it allows to teriminate the loop

```python
b=[1,2,5,90,'divya','arunachalam',7+8j,True,False]

for i in b:
    if type(i)==str:
        break
    print(i)
```

```python
a ="we are working healthcareprofessionals"


for i in a:
    if i=="g":
        break
    print(i)
```

# Continue
  -  when specific condition met, it allows to skip the somepart of statement and control transfered to next statement 

```python
a ="we are working healthcareprofessionals"


for i in a:
    if i=="g":
        continue
    print(i)
```

```python
b=[1,2,5,90,'divya','arunachalam',7+8j,True,False]

for i in b:
    if type(i)==str:
        continue
    print(i)
```

```python
b=[1,2,5,90,'divya','arunachalam',7+8j,True,False]

for i in b:
    if type(i)==int:
        continue
    print(i)
```
